```python
def convert(obj):
    genres = []
    for i in ast.literal_eval(obj):
        genres.append(i['name'])
    return genres
```

```python
print(convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'))
```

```python
# the list of dictionaries is actually a string
# to convert into a list
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(3)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['cast'][0]
```

```python
def convert3(obj):
    genres = []
    counter = 0 
    for i in ast.literal_eval(obj):
        if counter != 3:
            genres.append(i['name'])
            counter+=1
        else:
            break
    return genres
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(3)
```

```python
for i in ast.literal_eval(movies['crew'][0]):
    if i['job'] == 'Director':
        print(i['name'])
```

```python
ast.literal_eval(movies['crew'][0])
```

```python
def fetch_director(obj):
    director = []
    for i in ast.literal_eval(obj):
        if i['job']  == 'Director':
            director.append(i['name'])
            break
    return director
```

```python
fetch_director(movies['crew'][0])
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(3)
```

```python
# overview column is a string---> convert to a list
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head(3)
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
# tags column---> concatenate all the columns(except id and title)
```

```python
movies['tags'] = movies['overview'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['id', 'title', 'tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: ' '.join(x))
# convert the tags to a string
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
(cosine_similarity(vectors)).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
new_df[new_df['title'] == 'Batman Begins'].index[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
